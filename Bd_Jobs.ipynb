Import Libraries

```python
!pip install beautifulsoup4
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
##scraping the info of all IT job listed in bdjobs.com

URL = "https://jobs.bdjobs.com/jobsearch.asp?fcatId=8&icatId="
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

Create a Dataframe

```python
columns = ['Job Position', 'Company', 'Location', 'Education', 'Experience', 'Deadline']
df = pd.DataFrame(columns=columns)
df
```

Parse the HTML and Save the Job Info in the DF

```python
title = soup.find_all("div", class_='job-title-text')

company = soup.find_all("div", class_='comp-name-text')

location = soup.find_all("div", class_='locon-text')

education = soup.find_all("div", class_='edu-text')

experience = soup.find_all("div", class_='exp-text')

deadline = soup.find_all("div", class_='dead-text')


pattern = "[\d]{1,2}\ [A-Z][a-z]+\ [\d]{4}" ##for deadline

for idx,value in enumerate(zip(title,company,location,education,experience,deadline)):

  df.loc[len(df.index)] = [value[0].find('a').text.strip(), value[1].get_text().strip(), value[2].get_text().strip(), value[3].get_text().strip(), value[4].get_text().strip(), re.findall(pattern, value[5].get_text().strip())[0]]
```
