```python
!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
columns = ['Job Position', 'Company', 'Location', 'Education', 'Experience', 'Deadline']
df = pd.DataFrame(columns=columns)
df
```

```python
wd.get("https://jobs.bdjobs.com/jobsearch.asp?fcatId=8&icatId=")


title = wd.find_elements(By.CLASS_NAME, 'job-title-text')

company = wd.find_elements(By.CLASS_NAME, 'comp-name-text')

location = wd.find_elements(By.CLASS_NAME, 'locon-text')

education = wd.find_elements(By.CLASS_NAME, 'edu-text')

experience = wd.find_elements(By.CLASS_NAME, 'exp-text')

deadline = wd.find_elements(By.CLASS_NAME, 'dead-text')


pattern = "[\d]{1,2}\ [A-Z][a-z]+\ [\d]{4}" ##for deadline


for idx,value in enumerate(zip(title,company,location,education,experience,deadline)):

  df.loc[len(df.index)] = [value[0].get_attribute("textContent").strip(), value[1].get_attribute("textContent").strip(), value[2].get_attribute("textContent").strip(), value[3].get_attribute("textContent").strip(), value[4].get_attribute("textContent").strip(), re.findall(pattern, value[5].get_attribute("textContent").strip())[0]]

  
```
