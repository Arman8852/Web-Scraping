```python
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get("https://www.python.org")
search_bar = driver.find_element(By.CSS_SELECTOR, '[name="q"]')
search_bar.clear()
search_bar.send_keys("Computer")
search_bar.send_keys(Keys.RETURN)
#search_bar.submit()
#print(driver.current_url)

title_list = []
location_list = []
remarks_list = []
job_list = []

ul = driver.find_element(By.CSS_SELECTOR, '[class="list-recent-events menu"]')
title_pattern = '^Job\ -\ [A-Za-z].'
location_pattern =  '^Location:\ [A-Za-z].'
li = ul.find_elements(By.TAG_NAME, "li")

for idx,item in enumerate(li):  

  title = item.find_elements(By.TAG_NAME, "a")


  if re.search(title_pattern, title[0].get_attribute("textContent").strip()):

     title_list.append(re.split('- ',title[0].get_attribute("textContent").strip())[1])
  

  info = item.find_elements(By.TAG_NAME, "p")

  remarks = ''

  for i, v in enumerate(info):

      if re.search(location_pattern, v.get_attribute("textContent").strip()):


        location = re.split(': ',v.get_attribute("textContent").strip())[1]

        location_list.append(location)

      else:

        remarks = remarks +' '+ v.get_attribute("textContent").strip()

  remarks_list.append(remarks)

job_list = list(zip(title_list,location_list,remarks_list) )
```

```python
columns = ['Job Position', 'Location', 'Remarks']
df = pd.DataFrame(data=job_list, columns=columns)
df
```
